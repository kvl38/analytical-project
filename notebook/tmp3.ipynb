```python
def data_loader(link: str) -> pd.DataFrame:
    
    """
    Загружает данные из CSV-файла по указанному пути или URL.

    Параметры:
    ----------
    link : str
        Путь к локальному CSV-файлу или URL к удалённому источнику.

    Возвращает:
    ----------
    pd.DataFrame
        DataFrame с загруженными данными. 
        В случае ошибки возвращается пустой DataFrame, а сообщение об ошибке выводится в консоль.
    """
    
    try:
        df = pd.read_csv(link)
        return df
    
    except Exception as e:
        print(f"Произошла ошибка: {e}")
        return pd.DataFrame()
```

```python
def renaming_columns(df: pd.DataFrame) -> pd.DataFrame:
    
    """
    Переименовывает столбцы DataFrame из camelCase/PascalCase в snake_case.

    Параметры:
    -----------
    df : pd.DataFrame
        Исходный DataFrame с названиями столбцов в camelCase или PascalCase.

    Возвращает:
    -----------
    pd.DataFrame
        Новый DataFrame с переименованными столбцами в формате snake_case.
    """
    
    def to_snake_case(col_name: str) -> str:
        
        """
        Преобразует отдельное имя столбца, заменяя переходы от строчных букв к заглавным 
        на нижние подчёркивания и приводит строку к нижнему регистру.

        Параметры:
        -----------
        name : str
            Исходное имя столбца.

        Возвращает:
        -----------
        str
            Имя столбца, преобразованное в snake_case.
        """
        
        stage_1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', col_name)
        stage_2 = re.sub('([a-z0-9])([A-Z])', r'\1_\2', stage_1)
        return stage_2.lower()

    new_columns = {col: to_snake_case(col) for col in df.columns}
    return df.rename(columns=new_columns)
```

```python
def df_analysis(df: pd.DataFrame, df_name: str = '') -> None:
    
    """
    Выполняет базовый анализ переданного DataFrame и выводит ключевую информацию.

    Параметры:
    -----------
    df : pd.DataFrame
        Датафрейм для анализа.
    df_name : str, optional
        Имя датафрейма.

    Функция выводит:
    -----------
    - Название датафрейма.
    - Количество явных дубликатов.
    - Количество пропущенных значений по каждому столбцу.
    - Для каждого столбца — число уникальных значений.
    """
    
    print(f"Для датафрейма {df_name} характерно:")
    print('\nЧисло явных дубликатов: ', df.duplicated().sum(),
          '\n\nПропущенные значения:\n', df.isnull().sum(), sep='')
    
    for column in df.columns:
        print(f'\nВ столбце \'{column}\'', df[column].nunique(), 'уникальных значений')
```

```python
def graph_quant_feature(df: pd.DataFrame, col_name: str, col_title: str) -> None:
    
    """
    Строит гистограмму и диаграмму размаха, отображающие различия в количественном признаке
    в зависимости от значения целевой переменной.

    Параметры:
    ----------
    df : pd.DataFrame
        Исходный датафрейм с данными. Должен содержать столбец 'subscriber_left'.

    col_name : str
        Название количественного признака, для которого необходимо построить графики.

    col_title : str
        Заголовок, общий для обеих визуализаций.

    Возвращает:
    -----------
    None
        Функция не возвращает значения, а отображает графики и выводит описательную статистику.
    """

    target = 'subscriber_left'
    
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 8))
    
    bins = int(np.log2(len(df[col_name])) + 1)
    palette = {0: '#4C72B0', 1: '#DD8452'}

    sns.histplot(data=df, x=col_name, hue=target, bins=bins, kde=False,
                 palette=palette, ax=axes[0], multiple='stack', edgecolor='black')
    axes[0].set_title('Гистограмма по группам')
    axes[0].set_xlabel(f'Значения признака: {col_name}')
    axes[0].set_ylabel('Количество абонентов')
    axes[0].legend(title='Абонент', labels=['Ушел (1)', 'Остался (0)'])
    axes[0].grid(True)

    sns.boxplot(data=df, x=col_name, y=target, palette=palette, ax=axes[1], orient='h')
    axes[1].set_title('Диаграмма размаха по группам')
    axes[1].set_xlabel(f'Значения признака: {col_name}')
    axes[1].set_ylabel('Ушел абонент')

    plt.suptitle(col_title)
    plt.tight_layout()
    plt.show()

    print(f'Описательная статистика по признаку "{col_name}":\n')
    display(df.groupby('subscriber_left')[col_name].describe())
```

```python
def graph_cat_feature(df: pd.DataFrame, col_name: str, col_title: str) -> None:
    """
    Строит наложенные столбчатые диаграммы для категориального признака
    в зависимости от целевой переменной 'subscriber_left'.

    Параметры:
    ----------
    df : pd.DataFrame
        Исходный датафрейм с данными. Должен содержать столбец 'subscriber_left'.

    col_name : str
        Название категориального признака.

    col_title : str
        Заголовок графика.

    Возвращает:
    -----------
    None
        Отображает график и печатает распределение значений.
    """

    target = 'subscriber_left'
    palette = {0: '#4C72B0', 1: '#DD8452'}

    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=col_name, hue=target, palette=palette, edgecolor='black', alpha=0.7, dodge=False)

    plt.title(col_title)
    plt.xlabel(f'Значения признака: {col_name}')
    plt.ylabel('Количество абонентов')
    plt.legend(title='Абонент', labels=['Остался (0)', 'Ушел (1)'])
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

    print(f'Распределение по категориям признака "{col_name}":')
    display(df.groupby([col_name, 'subscriber_left']).size().unstack(fill_value=0).T)
```

```python
def plot_corr_heatmap(df_corr: 'pd.DataFrame') -> None:
    
    """
    Строит heatmap корреляционной матрицы.

    Параметры:
    -----------
    df_corr : pd.DataFrame
        Матрица корреляций (df.corr() или phik_matrix).

    Возвращает:
    -----------
    None
        Функция отображает график и не возвращает значения.
    """
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(df_corr, annot=True, fmt=".2f", cmap='coolwarm')
    plt.title('Корреляционная матрица')
    plt.show()
```

## Объединение данных

### Объединение датафреймов

Перед объединением всех таблиц повторно рассмотрим их размерности

```python
print('contract_df:', contract_df.shape, 
      '\npersonal_df:', personal_df.shape,
      '\ninternet_df:', internet_df.shape,
      '\nphone_df:', phone_df.shape)
```

Из-за разной размерности датафреймов при объединении могут появиться пропущенные значения, которые потребуется обработать. При этом важно сохранить информацию по каждому договору абонентов, поэтому все остальные датафреймы будут присоединяться к contract_df

```python
# Объединяем все данные по столбцу 'customer_id'
merged_df = contract_df.merge(personal_df, on='customer_id', how='left') \
                       .merge(internet_df, on='customer_id', how='left') \
                       .merge(phone_df, on='customer_id', how='left')

# Проверяем корректность преобразований
merged_df.sample(10)
```

```python
# Рассмотрим размер объединенного датафрейма
merged_df.shape
```

ID абонентов больше не понадобится, так как не влияет на целевой признак

```python
# Удаляем столбец 'customer_id' из датафрейма
merged_df.drop('customer_id', axis=1, inplace=True)

# Проверяем корректность преобразований
merged_df.columns
```

### Обработка пропущенных значений

```python
# Изучим количество пропущенных значений
merged_df.isna().sum()
```

После объединения данных появились пропуски. Это могло произойти из-за системного сбоя при сохранении данных или из-за того, что не каждый абонент пользуется услугами телефонии или интернета, и информация о них отсутствует

```python
# Рассмотрим абонентов, у которых появились пропуски в столбцах, связанных с интернет-услугами и телефонией
merged_df[(merged_df['internet_service'].isna()) & (merged_df['multiple_lines'].isna())]['type'].count()
```

Таких абонентов не обнаружено, что означает: каждый клиент либо использует интернет, либо телефонную связь, либо обе услуги одновременно.

Выполним замену пропущенных значений: в столбцах *internet_service* и *multiple_lines* установим значение 'not_in_use', чтобы отразить, что абонент не пользуется соответствующей услугой; в столбцах *online_security, online_backup, device_protection, tech_support, streaming_tv и streaming_movies* установим значение 'No', что также указывает на отсутствие подключения этих дополнительных услуг

```python
# Заменяем пропуски в столбцах 'internet_service' и 'multiple_lines' на 'not_in_use'
merged_df['internet_service'] = merged_df['internet_service'].fillna('not_in_use')
merged_df['multiple_lines'] = merged_df['multiple_lines'].fillna('not_in_use')
```

```python
# Заполняем пропуски значением 'No'
internet_columns = ['online_security', 'online_backup', 'device_protection',
                    'tech_support', 'streaming_tv', 'streaming_movies']

merged_df[internet_columns] = merged_df[internet_columns].fillna('No')
```

```python
# Найдем число полных дубликатов после заполнения пропусков
merged_df.duplicated().sum()
```

```python
# Рассмотрим эти дубликаты
merged_df[merged_df.duplicated(keep=False)].sort_values(by='type')
```

```python
# Дубликатов немного, и их удаление способствет улучшению обучения ml-модели
merged_df = merged_df.drop_duplicates()

# Проверяем корректность преобразований
merged_df.duplicated().sum()
```

```python
# Выводим основную информацию нового датафрейма
merged_df.info()
```

- В новом датафрейме представлена информация о 7039 абонентах, описанных 19-ю признаками.
- Пропущенные значения отсутствуют.
- При заполнении недостающих данных в столбцах 'internet_service' и 'multiple_lines' была добавлена новая категория — 'not_in_use', означающая отсутствие подключения абонента к данной услуге. В столбцах 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv' и 'streaming_movies' установлено значение 'No', что также указывает на отсутствие подключения этих дополнительных услуг.

## Подготовка данных

```python
# Изменим тип данных object на category, чтобы корректно обработать эти признаки при обучении моделей
for column in merged_df.select_dtypes(include='object').columns:
    merged_df[column] = merged_df[column].astype('category')
    
# Проверяем корректность преобразований
merged_df.dtypes
```

```python
# Разделяем данные на обучающую и тестовую выборки с учетом дисбаланс классов
X = merged_df.drop('subscriber_left', axis=1)
y = merged_df['subscriber_left']

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=TEST_SIZE, 
                                                    random_state=RANDOM_STATE,
                                                    stratify=y)

# Рассмотрим размеры сформированных выборок
X_train.shape, X_test.shape
```

```python
# Создаём списки с названиями категориальных и непрерывных признаков для последующего кодирования и масштабирования
ohe_columns = list(merged_df.select_dtypes(include='category').columns)
num_columns = interval_columns
```

```python
# Создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
            [
                ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False), ohe_columns),
                ('num', MinMaxScaler(), num_columns)
            ],
            remainder='passthrough'
)
```

```python
# Изучим результат преобразования признаков
X_train_prepared = data_preprocessor.fit_transform(X_train)
X_train_prepared = pd.DataFrame(X_train_prepared, columns=data_preprocessor.get_feature_names_out())
X_train_prepared.head(3)
```

**Вывод по разделу:**
- Для более эффективной обработки категориальные признаки были переведены из типа *object* в *category*.
- С учётом дисбаланса классов данные были разделены на обучающую и тестовую выборки в соотношении 3:1.
- Признаки были разделены на *ohe_columns* и *num_columns* для их корректной обработки перед обучением моделей.
- Создан единый пайплайн для подготовки данных — data_preprocessor.
- Продемонстрирован пример датафрейма после обработки: категориальные признаки — методом OneHotEncoder, числовые — с помощью MinMaxScaler.

## Обучение ML-моделей

```python
# Вычисляем веса классов для обучения моделей
class_weights = compute_class_weight(class_weight='balanced',
                                     classes=np.unique(y_train),
                                     y=y_train)

class_weights_dict = dict(enumerate(class_weights))
class_weights_dict
```

```python
# Создаём итоговый пайплайн: подготовка данных и обучение модели
pipe_final = Pipeline([
                        ('preprocessor', data_preprocessor),
                        ('models', LogisticRegression(random_state=RANDOM_STATE))
                    ])

param_grid = [

        {
            'models': [LogisticRegression(random_state=RANDOM_STATE, max_iter=1000, penalty='l1', class_weight='balanced',)],
            'models__C': [0.01, 0.1, 1],
            'preprocessor__num': [StandardScaler(), MinMaxScaler()]
        },

        {
            'models': [RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced')],
            'models__n_estimators': [100, 200],
            'models__max_depth': [7, 10],
            'models__min_samples_split': [3, 5, 10],
            'preprocessor__num': ['passthrough']
        },

        {
            'models': [LGBMClassifier(random_state=RANDOM_STATE, class_weight='balanced')],
            'models__n_estimators': [250, 500],
            'models__learning_rate': [0.001 ,0.01],
            'models__max_depth': [3, 5],
            'preprocessor__num': ['passthrough']
        },

        {
            'models': [CatBoostClassifier(verbose=0, random_state=RANDOM_STATE, class_weights=class_weights_dict)],
            'models__iterations': [500, 1000],
            'models__depth': [3, 5, 7],
            'preprocessor__num': ['passthrough']
        }
    
]
```

```python
# Найдем лучшую модель, используя метрику 'roc_auc'
start_time = time.time()

grid_search = GridSearchCV(pipe_final, 
                           param_grid,
                           cv=CV,
                           scoring='roc_auc',
                           n_jobs=-1) 

grid_search.fit(X_train, y_train)

end_time = time.time()

best_model = grid_search.best_estimator_
print('Лучшая модель и её параметры:\n', best_model)
print(f"\nВремя поиска лучшей модели: {end_time - start_time:.2f} секунд")
```

```python
# Выводим получившуюся метрику модели на тренировочной выборке
print ('Метрика лучшей модели на тренировочной выборке:', round(grid_search.best_score_, 5))
```

```python
# Получаем важность и имена признаков
feature_importances = best_model.named_steps['models'].get_feature_importance()
feature_names = best_model.named_steps['preprocessor'].get_feature_names_out()

feature_importances_df = pd.DataFrame({'feature': feature_names,
                                       'importance': feature_importances})

# Сортируем по возрастанию важности признаков
feature_importances_df = feature_importances_df.sort_values(by='importance')

# Строим график
plt.figure(figsize=(12, 8))
plt.barh(feature_importances_df['feature'], feature_importances_df['importance'])
plt.xlabel('Важность признака')
plt.ylabel('Название признака')
plt.title('Важность признаков для CatBoost')
plt.tight_layout()
plt.show()
```

Наибольшее влияние на целевой признак оказывает количество дней с момента заключения договора *num_of_days*. На втором месте по значимости — ежемесячные расходы клиента *monthly_charges*, а на третьем — тип оплаты *type*, отражающий, производится ли оплата ежемесячно или раз в год или два.

**Вывод по разделу:**
- Для корректной работы моделей были рассчитаны веса классов.
- С использованием пайплайна для подготовки данных, обучения моделей и подбора гиперпараметров с помощью GridSearchCV среди четырёх алгоритмов машинного обучения (LogisticRegression, RandomForestClassifier, LGBMClassifier, CatBoostClassifier) была выбрана модель, показавшая наилучший результат по метрике ROC-AUC (0.87926) на обучающей выборке — CatBoostClassifier.
- На целевой признак наибольшее влияние оказывают: количество дней с момента заключения договора *num_of_days*, ежемесячные расходы клиента *monthly_charges*, тип оплаты — раз в год или два *type_two_year, type_one_year*, а также наличие у абонента супруга или супруги *partner_Yes*.
